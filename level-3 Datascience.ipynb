```python
#most used tag
book_tag
```

```python
#most used tag with tag_id
tag_counts = book_tag.groupby('tag_id')['count'].sum()
max_count_tag_id = tag_counts.idxmax()
min_tag_id = tag_counts[tag_counts == tag_counts.max()].idxmin()

print(f"The most frequently used tag_id is {min_tag_id}.")
```

```python
#book with most counts of tags
merged_data = pd.merge(book_tag, book, on='goodreads_book_id')
book_tag_counts = merged_data.groupby('title')['count'].sum()
most_tagged_book = book_tag_counts.idxmax()

print(f"Book with the most number of counts of tags: {most_tagged_book}")
```

```python
#ploting bar chart
tag_counts = book_tag.groupby('tag_id')['count'].sum()
sorted_tags = tag_counts.sort_values(ascending=False)
top_20_tags = sorted_tags.head(20)


plt.figure(figsize=(10, 6))
top_20_tags.plot(kind='bar', color='skyblue')
plt.title('Top 20 Tags by User Records')
plt.xlabel('Tag ID')
plt.ylabel('User Records')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
